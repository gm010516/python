###### 문제1-2) 답안 작성

키워드 인수: 함수에 입력되는 매개변수의 변수명을 사용하여 함수의 인수를 지정하는 방법

디폴트 인수: 매개변수에 기본값을 지정하여 사용하고 아무런 값도 인수로 넘기지 않으면 저장된 기본값을 사용하는 방식

키워드 가변인수: 매개변수의 이름을 따로 지정하지 않고 입력하는 방법으로 이전 가변 인수와는 달리 *를 2개 사용하여 함수의 매개변수를 표시함

    입력된 값은 튜플 자료형이 아닌 딕셔너리 자료형으로 사용할 수 있음
    키워드 가변 인수는 반드시 모든 매개변수의 맨 마지막, 즉 가변 인수 다음에 선언되어야함
    변수명은 중요하지 않지만 반드시 **붙여야함

가변인수의 예시)

```python
def asterisk_test_1(a,b,*args):
    return a+b+sum(args)
print(asterisk_test_1(1,2,3,4,5))

def asterisk_test_2(a,b,*args):
    print(args)
print(asterisk_test_2(1,2,3,4,5))

def asterisk_test_3(*args):
    x,y,*z=args 
    return x,y,z
print(asterisk_test_3(3,4,5,6))
```

####  문제 1-3) 예외처리에서 사용되는 3가지 구문에 대하여 설명하시오(1점)
1)try~except 구문

```python
for i in range(10):
    try:
        print(10/i)
    except ZeroDivisionError as e:
        print(e)
        print("Not divided by 0")
```

2)try~ except~else구문

```python
for i in range(10):
    try:
        result=10/i
    except ZeroDivisionError:
        print("Not divided by 0")
    else:
        print(10/i)
```

3)try~ except~finally구문

```python
try:
    for i in range(1,10):
        result=10/i
        print(result)
except ZeroDivisionError:
        print("not divided by 0")
finally:
        print("종료되었습니다.")
```

### 문제2) 함수, 예외처리, 파일처리에 대하여 각각의 문제에 대하여 답하시오
#### 문제2-1) 정상적인 성적은 그대로 출력, 오류성적은 출력하지 않고, 정상처리와 오류처리 된 것의 개수를 구하는 프로그램을 작성하라. (3점)

* 정상점수 범위 : 0~100
* 함수호출 : scoreproc(99, 101, 86), scoreproc(92, 86, 68, -1, 56)

* 실행결과) 

```python
# 문제2-1) 답안작성 
def scoreproc(*args):
    error=0
    valid=0
    for i in args:
        if 0<=i<=100:
            try:
                print(i)
                valid=valid+1
               
            except:
                 error=error+1
                
    #함수 프로그램 작성...

    
    
    print("정상처리: ", valid)
    print("오류처리: ", error)
    print("성적 처리 끝")

scoreproc(99, 101, 86)
scoreproc(92, 86, 68, -1, 56)
```

#### 문제2-2) 예외 처리 구문을 사용하여 사용자가 올바른 값(정수값 0~100 )을 입력할 때 까지 다시 받도록  프로그램을 수정하시오(3점)
* 처리사항1 : 에러가 발생하면 실행결과와 같이 몇번째 에러인지 표시하시오(에러 발생시만 계속 증가함)
* 처리사항2 : 정상으로 점수가 입력되면 화면에 "정상, 입력한 점수:80" 출력 후 프로그램 종료 할 것
* 입력사항 : 100점, -10, 80
* 실행결과) 

```python
# 문제2-2) 답안 작성 
# icnt변수를 사용하여 몇번째 오류인지 처리

icnt = 0
while True:
    
    try:
        in_num = int(input("점수 0~100 사이의 값을 입력하세요 : "))
        if in_num >= 0 and in_num <= 100:
            print("정상, 입력한 점수 : ", in_num)
        
        else:
            icnt=icnt+1
            print(icnt,"번째 오류, 점수의 범위는 0 ~ 100 사이의 값입니다.")
    except ValueError:
         # 예외처리에서 정수 이외의 값이 들어온 경우 에러 메시지 처리
        icnt=icnt+1
        print(icnt,"번째 오류, 오류 정수만 입력하세요~~")
     # 정수 범위가 아닌 경우 에러 메시지 처리
```

#### 문제3-3) 연합뉴스 최신기사의 제목, URL주소, 요약을 추출하는 프로그램을 작성하시오(4점)

* 분석 주소 : https://www.yna.co.kr/news/1
* 최신기사는 실시간 반영되므로 변경된 값으로 최신기사의 제목, URL주소. 요약 데이터 추출



* 실행 결과 : 

```python
# 문제3-3) 답안작성
import requests
from bs4 import BeautifulSoup

url = "https://www.yna.co.kr/news/1"
req = requests.get(url)

html = req.content
soup = BeautifulSoup(html,'html.parser')
 

if req.status_code == 200 : 

    tdlist = soup.select("div.list-type038>ul.list>li")

    for textname in tdlist:
        mtitle = textname.select_one("a>strong.tit-news").text.strip()
        mlink = textname.select_one("div.news-con>a").get("href")
        mcon = textname.select_one("div.news-con>p").text.strip()
        print("기사제목 : ", mtitle)
        print("URL주소 : https:"+mlink)
        print("기사요약 : ", mcon)
        print("="*100)
else:
    print("연합뉴스 사이트에서 정상적으로 호출되지 않았습니다.")
```

### 문제4) 위의 문제3-3) 프로그램을 수정하여 화면에 출력된 기사제목, URL주소, 기사요약을 텍스트 파일로 저장하는 프로그램을 작성하시오(10점)

* 텍스트 파일명 : news.txt
* 최신기사는 실시간 반영되므로 변경된 값으로 최신기사의 제목, URL주소. 요약 데이터 추출


* 실행 결과 : 

```python
# 문제4) 답안작성
import requests
from bs4 import BeautifulSoup


url = "https://www.yna.co.kr/news/1"
req = requests.get(url)

html = req.content
soup = BeautifulSoup(html,'html.parser')
 

if req.status_code == 200 : 

    tdlist = soup.select("div.list-type038>ul.list>li")

    for textname in tdlist:
        mtitle = textname.select_one("a>strong.tit-news").text.strip()
        mlink = textname.select_one("div.news-con>a").get("href")
        mcon = textname.select_one("div.news-con>p").text.strip()
        print("기사제목 : ", mtitle)
        print("URL주소 : https:"+mlink)
        print("기사요약 : ", mcon)
        print("="*100)
        f=open("news.txt","w")
        f.write(textname)
        f.close()

else:
    print("연합뉴스 사이트에서 정상적으로 호출되지 않았습니다.")

```

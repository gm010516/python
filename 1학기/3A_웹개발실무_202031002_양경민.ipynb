* 조건1) 더하기(A, a),   빼기(S, s),   곱하기(M, m),   나누기(D, d) <br>

* 조건4) 출력결과와 같이 정답이면 연산결과 출력, 오답인 경우 출력결과와 같이 메시지와 정답 결과 출력


```python
# 문제3-3) 답안작성
j=input("원하는 연산 선택:")
a=int(input("첫 번째 숫자 입력:"))
b=int(input("두 번째 숫자 입력:"))
c=(input("연산 결과 맞추기=>"))

if(j=="a" or j=="A"):
    if(c==a+b):
        print(a,"+",b,"=",a+b)
    else:
        print(c,"는 오답입니다.")
elif(j=="s" or j=="S"):
    if(c==a-b):
        print(a,"-",b,"=",a-b)
    else:
        print(c,"는 오답입니다.")
elif(j=="m"or j=="M"):
    if(c==a*b):
        print(a,"x",b,"=",a*b)
    else:
        print(c,"는 오답입니다.")
elif(j=="s"or j=="S"):
    if(c==a/b):
        print(a,"%",b,"=",a/b)
    else:
        print(c,"는 오답입니다.")
elif(b==0):
    print("0으로 나눌 수없습니다.")
else:
    print()
```

### 문제4) 반복문을 사용하여 각각의 문제에 대하여 프로그램 처리 하시오
#### 문제4-1) for 문으로 1 에서 10 까지 출력하되 3 의 배수를 건너뛰는 프로그램을 작성하라


```python
# 문제4-1) 답안작성

for i in range(1,11,1):
    if(i==10):
        print(i,end="")
    elif(i%3==0):
        continue
    else:
        print(i,end="/")
```

#### 문제4-2) for 문으로 시작수와 끝수를 입력 받아 범위 내의 솔수, 짝수의 합을 표시하는 프로그램을 작성하시오(3점)


```python
#문제4-2) 답안작성 
시작=int(input("시작수:"))
끝수=int(input("끝수:"))
sum1=0
sum2=0
for i in range(시작,끝수+1,1):
    if(i%2==0):
        sum1=sum1+i  
    else:
        sum2=sum2+i
print("홀수합=",sum2)
print("짝수합=",sum1)
```

#### 문제4-3) while 문으로 다음과 같은 결과가 나오도록 프로그램을 작성하시오(3점)


```python
#문제 4-3) 답안작성
i=1
while i<=5:
    print("----+-----",end="")
    i=i+1
```

#### 문제4-4) 1 부터 10 사이의 수치 중 2 의 배수나 3 의 배수를 출력하고 합을 구하는 프로그램을 작성하시오(3점)


```python
#문제4-4) 답안작성

sum1=0
sum2=0
for i in range(1,11,1):
    if(i%2==0):
        print("2의 배수:",i)
        sum1=sum1+i
    if(i%3==0):
        print("3의 배수:",i)
        sum2=sum2+i

print("\n2의 배수의 합:",sum1)
print("3의 배수의 합:",sum2)
```

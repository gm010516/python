* **if문 ex_08**
    * 문제) 아래 조건에 맞게 프로그램을 작성하라.
        * ① 연산문자를 입력 받고 숫자 2개를 입력받아 연산문자에 맞게 연산이 되도록 작성. 
             (더하기(A), 빼기(S), 곱하기(M), 나누기(D)   
        * ② 대문자나 소문자(a, s, m, d)를 입력해도 4칙연산이 되도록 수정.
        * ③ A,S,M,D (a,s,m,d)를 입력하지 않았을 경우“A, S, M, D중 하나를 넣으세요.”라는 메시지 출력.
        * ④ 두 번째 숫자에 0을 입력하면 “0으로 나눌 수 없습니다.”라는 메시지 출력.


```python
ch=input("더하기(A),빼기(S),곱하기(M),나누기(D)\n 원하는 연산선택:")
ch=ch.lower()
a = int(input("첫 번째 숫자입력:"))
b = int(input("두 번째 숫자입력:"))

if(ch=="a"): ##더하기
    print(str(a+b))
    
elif(ch=="s"): ##빼기
    print(str(a-b))

elif(ch=="m"): ##곱하기
    print(str(a*b))
    
elif(ch=="d"): ##나누기
    if(b==0):
        print("0으로 나눌 수 없습니다.")
    else:
        print(str(a/b))
else:
    print("a,s,m,d 중 하나만 입력하세요!")
    

"""
빼기에서 해결하는 다른 방법
    if(a>b):
        print(str(a-b))
    else:
        print(str(b-a))"""
        
```

* **if문 ex_09**
    * 문제) 컴퓨터와 가위, 바위, 보 게임 프로그램을 작성하시오 
            컴퓨터는 랜덤 사용함


```python
import random;
dice = ["가위", "바위", "보"]
computer = random.choice(dice)

player = input("(가위, 바위, 보) 중에서 하나를 입력 하세요: ");


number = random.randint(0,2);
if (number == 0):
    computer = "가위"
elif (number == 1):
    computer = "바위"
else:
    computer = "보"

print("사용자: ", player, "컴퓨터: ", computer)

if(player==computer):
    print("비겼음!")
elif(player=="바위"):
    if(computer=="보"):
        print("컴퓨터가 이겼음!")
    else:
        print("사용자가 이겼음!")
elif(player=="보"):
    if(computer=="바위"):
        print("사용자가 이겼음!")
    else:
        print("컴퓨터가 이겼음!")
elif(player=="가위"):
    if(computer=="바위"):
        print("컴퓨터가 이겼음!")
    else:
        print("사용자가 이겼음!")
```

### 3.3 리스트의 이해 (p78-89)
* 하나의 자료형으로만 저장하지 않고, 정수형, 실수형, 문자형 등의 다양한 자료형을 포함 할 수 있다.

#### 3.3.1 인덱싱(Indexing) (p. 79)
* 인덱싱: 리스트에 있는 값에 접근하기 위해, 이 값의 상대적인 주소를 사용하는 것
* len() 함수 : 리스트 안에 있는 값의 개수 반환

```python
colors = ['red', 'blue', 'green']
print(colors[0])

print(colors[1])

print(colors[2])

print(len(colors)) ##length의 약자-인덱스의 갯수를 알 수 있음

print(colors)

print(colors[-1])##리버스 인덱스
```

#### 3.3.2. 슬라이싱(slicing) (p. 80)
* 슬라이싱: 리스트의 인덱스를 사용하여 전체 리스트에서 일부를 잘라내어 반환
* 슬라이싱 기번 문법 : **변수명[시작인덱스:마지막인덱스]**

* cities[0:5] ==> 인덱스 0부터 (5-1)까지(서울, 부산, 인천, 대구, 대전)
* cities[:5] ==> 인덱스 시작부터 (5-1)까지(서울, 부산, 인천, 대구, 대전)
* cities[1:] ==> 인덱스 0부터 마지막 까지(부산, 인천, 대구, 대전, 광주, 울산, 수원)

```python
# 시작인덱스, 마지막 인덱스 사용하여 출력
cities = ['서울', '부산', '인천', '대구', '대전', '광주', '울산', '수원']
print(cities[0:4])
print(cities[0:5])
```

```python
# 마지막 인덱스값만 넣어서 출력
print(cities[:5])
```

```python
# 시작 인덱스값만 넣어서 출력
print(cities[3:])
```

#### 3.3.3 리버스 인덱스(reverse index) (p.81)
* 리스트에는 인덱스를 마지막 값부터 시작하는 리버스 인덱스 기능이 있다.
* 리버스 인덱스 : 

```python
# 리버스 인덱스 사용하여 마지막 값 출력
print(cities[-1])
```

```python
# 리버스 인덱스 사용하여 2개 출력
print(cities[-2:])
print(cities[-2:-1])
```

```python
# 리버스 인덱스 사용하여 "서울"~"대전"까지 출력
print(cities[-8:-3])
```

#### 3.3.4 인덱스 범위 넘어가는 슬라이싱 (p.82)
* 인덱스를 따로 넣지 않고 print(cities[:])과 같이 콜론(:)을 넣으면 cities 변수의 모든 값을 다 반환
* 슬라이싱에서는 인덱스를 넘어서거나 입력하지 않더라도 자동으로 시작 인덱스와 마지막 인덱스로 지정

```python
# 시작과 마자막 값 생략시 
print(cities[:])
print(cities)
```

```python
# 인덱스 범위를 넘어가는 경우
print(cities[:50])
###print(cities[-50])==에러
```

#### 3.3.5 증가값 (p.82)
* 슬라이싱에서는 시작 인덱스와 마지막 인덱스 외에 마지막 자리에 증가값을 넣을 수 있다.
* 증가값은 한 번에 건너뛰는 값의 개수이다.
* **변수명[시작 인덱스:마지막 인덱스:증가값]**

```python
# 2개씩 띄어서 출력 하는 경우 증가값 사용
print(cities[::2])
```

```python
# 리스트를 역순으로 출력 하는 경우
print(cities[::-1])
```

#### 3.4 리스트의 연산 (p.83-84)
* 리스트 연산 : **덧셈 연산, 곱셈 연산, in 연산**
* 덧셈연산 : 덧셈 연산을 하더라도 따로 어딘가 변수에 할당해 주지 않으면 기존 변수는 변화가 없다.
* 곱셈연산 : 리스트의 곱셈 연산은 기준 리스트에 n을 곱했을 때, 같은 리스트를 n배만큼 늘려 준다.
* in 연산  : 포함 여부를 확인하는 연산으로, 하나의 값이 해당 리스트에 들어 있는지 확인할 수 있다.

```python
# 리스트의 덥셈 연산
color1 = ['red', 'blue', 'green']
color2 = ['orange', 'black', 'white']
print(color1+color2)


t_color=color1+color2
print(t_color)
```

```python
# color1 리스트의 곱셈 연산
colors=color1*2
print(colors)
```

```python
# 리스트에 값의 포함 유무 판별(in 연산자 사용)/공백, 대소문자가 영향을 줌
"green" in colors
```

```python
# 리스트에 값의 포함 유무 판별(in 연산자 사용)/.lower()->전부 소문자로 바꾸기
"Green".lower() in colors
```

### 3.4 리스트의 메모리 관리 방식 (p90-95)
* 파이썬은 리스트를 저장할 때 값 자체가 아니라, 값이 위치한 메모리 주소(reference)를 저장한다.
* ==은 값을 비교하는 연산이고, is는 메모리의 주소를 비교하는 연산이다.            
* 리스트 메모리 관리 방식 

```python
# 값이 -5~256까지 인경우와 그 이외의 값인 경우 차이 확인, == 과 is의 차이점 출력
a = 250
b = 250

print("a값:",a, " id:",id(a))
print("b값:",b, " id:",id(b))

print(a==b)
print(a is b)
```

* 리스트의 메모리 저장 : 리스트는 기본적으로 값을 연속으로 저장하는 것이 아니라, 값이 있는 주소를 저장하는 방식이다.


```python
# 리스트는 다양한 형태의 변수가 들어 갈 수 있다.
a = ['color', 1, 10.5]
color = ['red', 'blue', 'green']
a[0] = color

print(a)
```

```python
# b와 a 변수를 각각 다른 값으로 선언 한 후 b에 a를 할당
a = [5,4,3,2,1]
b = [1,2,3,4,5]
b = a

print(b)
```

```python
a.sort()
print(b, id(b))
print(a, id(a))
```

* 리스트의 저장 방식 

```python
b = [6,7,8,9,10]
print(a, id(a))
print(b, id(b))
```

```python
a.reverse()
print(a, id(a))
```

* **반복문 ex_01**
    * 문제) 반복문으로 1에서 200까지 3의 배수의 합을 출력하는 프로그램을 작성하라.
            (while문, for문으로 각각 작성하시오)


```python
# while문 방법 1
sum=0
i=0
while i<=200:
    sum=i+sum
    i=i+3

print(sum)
```

```python
# while문 방법 2
i= 0
sum=0
while i<=200:
    if(i%3==0):
        sum=i+sum
        i=i+1
    else:
        i=i+1
print(sum)
```

```python
# for문
sum = 0
for i in range(0,201,3):
    sum=i+sum
print(sum)
```

* **반복문 ex_02**
    * 문제) 시작수와 끝수를 입력 받아 범위 내의 합을 표시하는 프로그램을 작성하라.
            (while문, for문으로 각각 작성하시오)

```python
# while문

start=int(input("시작수:"))
end=int(input("끝수:"))
i=start
sum=0
while i<=end:
    sum=i+sum
    i=i+1

print("sum=",sum)
```

```python
# for문

start=int(input("시작수:"))
end=int(input("끝수:"))
i=start
sum=0
for i in range(start,end+1,1):
    sum=i+sum


print("sum=",sum)
```

* **반복문 ex_05**
    * 문제) 아래 예제를 수정하여 정상적인 성적은 그대로 출력하고, 
            정상처리와 오류처리된 것의 개수를 구하는 프로그램을 작성하라. (정상점수:0~100)


```python
score = [ 92, 86, 68, -1, 56]
valid = 0
error = 0


for s in score:
    if s<0 or s>100:
        error=error+1
    else:
        valid=valid+1
        print(s)

print("정상=",valid)
print("오류=",error)
print("성적 처리 끝")
```

* **반복문 ex_08**

* **반복문 ex_10**
    * 문제) 문자열 address 문자열의 내장 목록에서 __가 포함되지 않은 함수를 화면에 출력하는 프로그램을 작성하라

```python
# dir을 사용하여 내장 목록 확인하기
address = "서울시 종로구 필운대로 1길, Pirundae-ro 1-gil, Jongno-gu, Seoul, Republic of Korea "
print(dir(address))
```

```python
# __개가 있는 스페셜 함수를 제외한 함수 목록 출력
print_func = []

for p_func in dir(address):
    print(p_func, end=",")
```

```python
# __개가 있는 스페셜 함수를 제외한 함수 목록 출력 완성 소스

print_func = []
for p_func in dir(address):
    if not p_func.startswith("__"):
        print_func.append(p_func) 
print(print_func)

```

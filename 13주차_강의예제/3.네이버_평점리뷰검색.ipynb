```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
user_url = "https://movie.naver.com/movie/point/af/list.naver?&page=2"
res = requests.get(user_url)

print(res.status_code)
```

```python
soup = bs(res.content, "html.parser")

#tag = "#cbody>#old_content>table.list_netizen>colgroup>tbody>tr"
tag = "#old_content>table.list_netizen>colgroup>tbody>tr"
values = soup.select(tag)

print(values)
```

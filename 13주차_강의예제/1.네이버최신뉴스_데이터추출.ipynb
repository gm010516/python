```python
#    
#   :   ''..4 '' 7
#   : kakaoWrap>#kakaoContent>div>div.head_view>h3.tit_view

from bs4 import BeautifulSoup
import requests as req

url = "https://entertain.naver.com/now#sid=106&date=2022-05-30&page=1"
html = req.get(url) 
soup = BeautifulSoup(html.content, "html.parser")

tag = "#content>div.ct_wrp>div.sub_cont>div.left_cont>ul>li"
values = soup.select(tag)

print(values)
```

```python
# a  href  
for v in values:
    temp = v.find("div").find("a")
    print("https://entertain.naver.com/"+temp.attrs["href"], temp.text)
```

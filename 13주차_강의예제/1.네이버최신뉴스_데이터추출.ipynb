### << 네이버 최신 뉴스 데이터 추출 하기 실습 >>
* 사이트 주소 : https://entertain.naver.com/now#sid=106&date=2022-05-30&page=2
* HTML 파싱에 사용되는 모듈 : from bs4 import BeautifulSoup
* 웹사이트 불러오는 모듈(2개중 사용) : import urllib.request as req, import requests as req 


```python
# 다음 뉴스의 타이틀명 추출하기
# 추출 데이터 : 경기지역 기초단체장 '리턴매치'..4년만에 '복수혈전' 7곳
# 핵심 키워드 : kakaoWrap>#kakaoContent>div>div.head_view>h3.tit_view

from bs4 import BeautifulSoup
import requests as req

url = "https://entertain.naver.com/now#sid=106&date=2022-05-30&page=1"
html = req.get(url) 
soup = BeautifulSoup(html.content, "html.parser")

tag = "#content>div.ct_wrp>div.sub_cont>div.left_cont>ul>li"
values = soup.select(tag)

print(values)
```

```python
# a 안의 href 특성 읽기
for v in values:
    temp = v.find("div").find("a")
    print("https://entertain.naver.com/"+temp.attrs["href"], temp.text)
```

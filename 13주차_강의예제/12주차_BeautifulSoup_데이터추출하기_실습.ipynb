##### <span style="color:red"> bs4(BeautifulSoup)  모듈</span>
* BeautifulSoup은 사용하여 웹페이지의 요청된 결과 값인 HTML에서 데이터를 쉽게 추출 할 수 있게 하는 모듈
* 많이 사용되는 함수는 find(), find_all(), select()
  * find() 함수 : HTML 태그에서 검색조건에 맞는 첫번째 태그 값만 검색하여 가져온다.
  * find_all()함수 : 해당 태그가 여러개 있을 경우 한꺼번에 데이터를 모두 가져 온다.
  * select_one() 함수 : HTML 태그에서 검색조건에 맞는 첫번째 태그 값만 검색하여 가져온다.
  * select()함수 : 필요한 데이터를 확인하고, 조건에 맞는 검색 태그의 전체 데이터를 추출하고자 할 때 select()함수와 find_all()함수를 사용하며 결과 값은 리스트 형식
      * select(“태그이름”) : 태그이름에 해당되는 태그를 검색하여 전체 데이터 추출
      * select(“.클래스명”) : 클래스명에 해당되는 태그를 검색하여 전체 데이터 추출
      * select(“부모태그>자식태그>자식태그”) : HTML태그 구조와 일치하는 태그를 검색하여 전체 데이터 추출
      * select(“부모태그.클래스명>자식태그.클래스명”) : 부모태그.클래스명과 자색태그.클래스명의 구조와 일치하는 태그를 검색하여 전체 데이터 추출
      * select(“#아이디명”)  : HTML태그의 아이디명에 해당되는 태그를 검색하여 전체 데이터 추출
      * select(“#아이디명>태그명.클래스명”) : HTML태그의 아이디명에 해당되고 자식의 태그명.클래스에 해당되는 태그를 검색하여 전체 데이터 추출
      * select(“태그명[속성]”) : 태그명의 속성에 해당되는 태그를 검색하여 전체 데이터 추출

```python
# 다음 뉴스의 URL주소와 타이틀명 추출 하기

from bs4 import BeautifulSoup
import urllib.request as req

def getParsing(url):
    html = req.urlopen(url)
    soup = BeautifulSoup(html, "html.parser")
    return soup

url = "https://news.daum.net/"
tag = "main.doc-main>section.inner-main>div.main-content>div.content-article>div>ul>li>div>div.cont_thumb>strong.tit_g>a"
soup1 = getParsing(url)

values = soup1.select(tag)

#print(values)
#print(type(values))
# a 안의 href 특성 읽기
for v in values:
    print(v.attrs["href"], v.text)
```

```python
# Step1
html = response.content
soup = BeautifulSoup(html, 'html.parser')
# print(soup) # 전체 HTML내용을 화면에 출력
print(soup.find('ul', 'presentation').find_all('li')) # UL태그의 클래스가 presentation의 하위 li태그만 화면에 출력
```

```python
# Step2
text1 = soup.select('ul.presentation > li') # Step1의 find, find_all함수대신 select함수 사용, 동일한 결과
print(text1)
```

```python
# Step3
urlsite = 'http://www.baewha.kr/Service/Sub/appl/03_01.asp'
response = rs.get(urlsite)
if response.status_code == 200:     #정상적인 경우만 수행되게 처리 함, urlsite값을 변경하여 확인 할 것
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    print(soup.find('ul', 'presentation').find_all('li'))
    text1 = soup.select('ul.presentation > li')
#    print(soup)
else : 
    print(response.status_code)
```

```python
# Step4 : 목록 전체의 텍스트만 출력되게 수정 (반복문 사용하여 처리)
for x in text1:
    print(x.get_text())
```

```python
# Step5 : URL주소, 이미지 파일명도 출력되게 수정
for x in text1:
    print(x.find("a")["href"].replace("\n", ""))  # find함수 사용하여 a태그의 href속성 값을 화면에 출력
    print(x.find("img")["src"])                   # find함수 사용하여 img태그의 scr속성 값을 화면에 출력
    print(x.get_text())                           # x 리스트의 값 전체의 텍스트를 화면에 출력
```

```python
# Step6 : 상세 내용의 태그 포함되어서  출력
for x in text1:
    print(x.find("a")["href"].replace("\n", ""))
    print(x.find("img")["src"])
#    print(x.get_text())
    print(x.select("ul.training > li"))
```

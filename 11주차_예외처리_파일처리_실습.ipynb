```python
# try ~ except 구문 연습 (p407)
for i in range(10):
    print(10 / i)
```

```python
# try ~ except 구문 연습 (p407)
for i in range(10):
    try:
        print(10 / i)
    except ZeroDivisionError as e:
        print(e)##에러의 종류를 알고 싶을 때
        print("에러발생, 0이외의 숫자 처리")
```

* **예외의 종류와 예외 에러 메시지**
    * 예외의 종류(p408)


* 예외 처리 구문 : **try ~ except ~ else문** (p409~410)


```python
# try ~ except ~ else 구문 연습 (p410)
for i in range(10):
    try:
        result=10/i
    except ZeroDivisionError as e:
        print(e)##에러의 종류를 알고 싶을 때
        print("에러발생, 0이외의 숫자 처리")
    else:
        print(result)
```

### 5.2 함수 심화 (p.169-171)
* **함수의 호출방식**
    * 값에 의한 호출 : 함수에 인수를 넘길때 겂만 넘김
    * 참조호출 : 함수에 인수를 넘길때 메모리 주소를 넘김
    

```python
# 값에 의한 호출(Call by Value) (p.169)
def f(x):
    y = x
    x=5 ##글로벌로 선언하면 전역변수로 사용 가능/지역변수
    print("x=5 적용됨", "메모리주소:", id(x))
    return y * y

x = 3
print(f(x))
print("x=3이 적용됨:",x, "메모리주소:",id(x))##별도의 메모리주소에 할당시킴/전역변수
```

```python
# 참조호출(Call by reference) (p.170-171)
def spam(eggs):
    eggs.append(1)
    print("eggs 함수호출:", eggs, "eggs 메모리주소:", id(eggs))
    eggs = [2,3]##리스트를 다른 값으로 재할당->지역변수로 변경됨
    print("eggs 재 할 당:", eggs, "eggs 메모리주소:", id(eggs))

ham = [0]
spam(ham)##리스트형으로 전달할 때는 포인터와 같이 메모리 할당 주소를 부여해줌
print("ham=", ham, "ham 메모리주소:", id(ham))
```

* **재귀함수** (p.175-177
* **재귀 함수(recursive function)**: 함수가 자기 자신을 다시 부르는 함수이다.

    * 팩토리얼(factorial) 함수이다. 정확히는 ‘n!’로 표시하면 n! = n × (n - 1)!로 선언할 수 있다. 자신의 숫자에서 1씩 빼면서 곱하는 형식이다. 보통은 점화식이라고 한다.

```python
# 재귀함수 예제 (p.176)
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)

print("결과값 :", factorial(int(input("Input Number for Factorial Calculation: "))))

# 계산과정 : 4*3*2*1 출력 되게 하려면?
```

#### 5.4.2 map() 함수
* **map(function, iterable)**
* **함수(function)**와 **반복 가능한(iterable) 자료형**을 입력 받음
* map 함수는 함수와 리스트를 인자로 받음
* 리스트로부터 원소를 하나씩 꺼내서 함수를 적용시키고, 그 결과를 새로운 리스트에 담아준다.
* 결과값이 리스트 형이므로 python3 버전에서 list 함수와 같이 사용

```python
# map() 함수 사용 : 리스트에 있는 항목을 반복해서 사용함
li = [1, 2, 3]
result = list(map(lambda i: i * i, li))
print(result)
```

#### 5.4.3 filter() 함수
* **filter(function, iterable)**
* 반복 가능한 자료형(iterable) 요소들을 function 함수에 하나씩 입력하여 **결과값이 참인 것만" 리스트로 반환
* 함수의 리턴 값은 참이나 거짓이어야 한다.
* 결과값이 리스트 형이므로 python3 버전에서 list 함수와 같이 사용

```python
# filter() 함수
list(filter(lambda x : x < 5, range(10)))
```

```python
# 결과값의 이해 (True:1, False:0)
list(filter(lambda x: x % 2, range(10)))
```

```python
# 양수만 반환하는 리스트 작성(일반 함수, filter함수, lambda 함수 비교)
li = [-2, -3, 5, 6]

# 일반 함수 사용
def pt(li):
    result = []
    for e in li:
        if e > 0:
            result.append(e)
        else:
            pass
    return result

print(pt(li))
```

```python
# filter 함수 사용
li = [-2, -3, 5, 6]
def pt1(x):
    return x > 0

list(filter(pt1, li))
```

```python
# filter + lambda 함수 사용
li = [-2, -3, 5, 6]

list(filter(lambda x: x > 0, li))
```

```python
list(filter(lambda x: x % 2==0, range(10)))
```

#### 5.4.4 reduce() 함수
* **reduce(function, 순서형자료)**
* python3부터 내장함수에서 빠짐
* 순서형 자료(문자열, 리스트, 튜플)의 원소들을 누적으로 함수에 적용

```python
# reduce()함수 : Python3에서는 외장 함수이므로 반드시 import해야 함
# 먼저 0과 1을 더하고, 그 결과에 2를 더하고, 거기다가 3을 더한 값을 리턴함
# 계산 순서 : (((0+1)+2)+3)
from functools import reduce 

reduce(lambda x, y: x + y, [0,1,2,3])
```

```python
#계산순서 : (e+(d+(c+(b+a)))
reduce(lambda x, y: y + x, 'abcde')
```

```python
# 일반적인 함수로 최소값 구하기
li = [-2, -3, 5, 6, 8, -6]

def max(li):
    default = 0
    for x in li:
        if default > x:
            default = x
    return default
max(li)
```

```python
# lambda 함수로 최소값 구하기
li = [-2, -3, 5, 6, 8, -6]

from functools import reduce
reduce(lambda x,y: x if x < y else y ,li)
```
